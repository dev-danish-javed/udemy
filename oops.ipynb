# OOPs in python

```python
class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def start(self):
        print(f'{self.make} started !!')

    def gear_shift(self):
        print(f'{self.make} gear shift !!')
```

```python
innova= Car('Toyota', 'Innova', '2021')
print(innova.make)
print(innova.model)
innova.start()
```

## Inheritance

```python
class Tesla(Car):
    def __init__(self, make, model, year, self_driving):
        super().__init__(make, model, year)
        self.self_driving = self_driving

    def is_self_driving(self):
        print(f'{self.make} {"supports" if self.self_driving else "doesn't support"} self driving !!')
```

```python
tesla_1 = Tesla('tesla','cyber truck', '2025', True)
tesla_1.start()
tesla_1.is_self_driving()
```

### Multiple Inheritance

```python
class Student:
    def __init__(self, first_name, last_name, stream):
        self.first_name = first_name
        self.last_name = last_name
        self.stream = stream

    def studies(self):
        print(f"Studnets don't study")

class Developer:
    def __init__(self, first_name, last_name, department):
        self.first_name = first_name
        self.last_name = last_name
        self.department = department

    def studies(self):
        print(f"Developers always study")
```

```python
class Freshman(Developer, Student):
    def __init__(self, first_name, last_name, department, stream):
        Developer.__init__(self=self, first_name=first_name, last_name=last_name, department=department)
        Student.__init__(self=self, first_name=first_name, last_name=last_name, stream=stream)
        self.first_name = first_name
        self.last_name = last_name
        self.department = department
        self.stream = stream

fresh = Freshman('danish', 'javed', 'web dev', 'CS')
fresh.studies()
```

# Polymorphism


```python
class Animal:
    def speak(self):
        print('I am animal')
        pass

class Dog(Animal):
    def speak(self):
        print('I am dog')

class Cat(Animal):
    def speak(self):
        print('I am cat')
```

This is polymorphism where the animal object will speak differently.


```python
def animal_speak(animal: Animal):
    animal.speak()
```

```python
dog = Dog()
cat = Cat()
animal_speak(dog)
animal_speak(cat)
```

# Abstraction

```python
from abc import ABC, abstractmethod


class Animal(ABC):
    @abstractmethod
    def speak(self):
        pass
    def eat(self):
        print('Animal is eating')

class Dog(Animal, ABC):
    @abstractmethod
    def speak(self):
        pass

class Puppy(Dog):
    def speak(self):
        print('I am puppy')
    def look_cute(self):
        print('I am looking so fukin cute')
```

```python
puppy:Dog = Puppy()
puppy.eat()
```

# Encapsulation

## Setter getter

```python
class Car:
    def __init__(self, make, model, year):
        self._year = year
        self.make = make
        self.model = model

    def __start(self):
        print(f'{self.make} started !!')

class Brezza(Car):
    def __init__(self, make, model, year):
        Car.__init__(self, make, model, year)
```

```python
brezza = Brezza("Suzuki","top", "2021")
brezza._Car__start()
```
