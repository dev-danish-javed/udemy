```python
from datetime import datetime


def checkEvenOdd():
    num = input("Enter the number")
    if not num.strip().lstrip('-').isdigit():
        print("Invalid input")
        return

    num = int(num)

    if num < 0:
        print("Negative number")
    elif num%2==0:
        print("Even")
    else :
        print("Odd")

checkEvenOdd()
```

```python
print(2, end="\t")
for x in range(3, 101, 2):
    isPrime = True
    for y in range(3, x):
        if x%y == 0:
            isPrime = False
            continue
    if isPrime:
        print(x, end="\t")
```

```python
def isPrime(num) -> bool :
    if num < 2:
        return False
    if num == 2:
        return True

    for i in range(2, num):
        if num % i == 0:
            return False
    return True

for i in range(1, 101):
    if isPrime(i):
        print(i, end="\t")
```

```python
def isPrime(num) -> bool :
    if num < 2:
        return False
    if num == 2:
        return True

    for i in range(2, int(num**0.5)+1):
        if num % i == 0:
            return False
    return True

for i in range(1, 101):
    if isPrime(i):
        print(i, end="\t")
```
