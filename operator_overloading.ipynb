# Operator overloading
Allows us to enable operations on our objects. This can also be done with functions but having this is also fine


```python
class Section:
    def __init__(self, students: list, lectures_count):
        self.students = students
        self.lectures_count = lectures_count

    def __add__(self, other):
        return Section(
            self.students + other.students,
            max(self.lectures_count, other.lectures_count)
        )

    def __str__(self):
        return f"{self.students}\t - \t{self.lectures_count}"
```

```python
section_a = Section([{"name":"Danish", "marks":78},{"name":"Rishabh", "marks":85}], 5)
section_b = Section([{"name":"Karan", "marks":87},{"name":"Ashu", "marks":95}], 3)
print(section_a + section_b)
```
