# Learn String Operations

```python
name = "Danish"
```

## Length of string

```python
len(name)
```

## Indexing
> `str[index]`

Index starts from 0
- Postive index : n_th element from start
- Negative index: n_th element from end

```python
# positive indexing - n_th element from start
print("name[0]:", name[0], end="\t\t")
print("name[3]:", name[3], end="\t\t")
print("name[-1]:", name[-1])
```

## Slicing
> `str[start:end:step]`

- start : Index where to start slicing from
- end : Index where to start slicing from
- step : Elements to skip


```python
count = '0123456789'
sliced_count = count[2:len(count):3]
print(sliced_count)

# ValueError: slice step cannot be zero
#sliced_count = count[2:len(count):0]
```

## Case opertaions


```python
name = "dAnIsH jaVEd"
print(name.lower())
print(name.upper())
print(name.title()) # first alphabet of each letter is upper case
print(name.capitalize()) # first alphabet of sentence is upper case
print(name.swapcase())
```

# Search and Check
- find : finds the index of substring in the string, returns -1 if the target string doesn't exists
- index : finds the index of substring in the string, throws exception if target string doesn't exists
- startswith : checks if string starts with the particular string
- endswith : verifies if the string ends with the particular string
- in : checks if the substring exists anywhere in the target string
- count :

```python
name = "Danish"
print(name.find("is"))
print(name.find("p"))

print(name.index("is"))
# print(name.index("p")) throws exception

print(name.startswith("Dan"))
print(name.startswith("sh"))

print(name.endswith("Dan"))
print(name.endswith("sh"))

print("Dan" in name)

banana = "banana"
print(banana.count("a"))
print(banana.count("n"))
print(banana.count("an"))
```

## Trim
- strip : removes space from both sides
- lstrip : removes space from start
- rstrip : removes space from end

```python
name = "    Danish    "
print(name)
print(name.strip())
print(name.rstrip())
print(name.lstrip())
```

## Replace and split
- replace : if the string to be replaced doesn't exists, no changes are made
- split : if the string to be replaced doesn't exists, an array with one element, the entire string, is returned

```python
testStr = "Hello World, how are you?"
print(testStr.replace("World", "Hania"))
print(testStr.replace("non-existance", "Hania"))
print(testStr.split(","))
print(testStr.split("World,"))
print(testStr.split("else"))
```

## Formatting
- f-string : f"My name is {name}"
- "I am {} and my age is {}".format(name, age)

```python
name = "Danish"
age = 28
print(f"Hi, I am {name} and my age is {age}.")
print("Hi, I am {} and my age is {}.".format(name, age))
```
